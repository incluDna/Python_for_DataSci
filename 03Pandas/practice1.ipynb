```python
import pandas as pd
import numpy as np
import random

#pd.Series (ทำได้หลายแบบ) -printออกมา จะโชว์indexด้วย ปกติจะเริ่มจาก 0
s1=pd.Series([random.randint(1,10) for i in range(4)], index=[1,2,3,4]) #บอกว่า4ตัวนี้ มีindexเป็น1,2,3,4 ไม่งั้นตัวแรกจะเริ่มที่ 0
s2=pd.Series(np.random.randint(low=1,high=10,size=4), index=[1,2,3,4])
print(s1)
print(s2)
```
