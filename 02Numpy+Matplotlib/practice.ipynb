````python
import random
import numpy as np #จะได้ไม่ต้องพิมยาว
import datetime

# for _ in range(จำนวนตัว) ถ้ามีแค่นี้ ไม่ต้องใส่ :
data1 =  [[random.randint(0,99) for j in range(10)] for i in range(10)]
print(data1)
count = 0

start_time = datetime.datetime.now() #เวลาเริ่ม
#ถ้าเป็นพวกเงื่อนไขต่อ ต้องใส่ :
for i in range(10):
  for j in range(10):
    if data1[i][j] > 50:
      count += 1
end_time = datetime.datetime.now()
print(count)
t = (end_time - start_time).microseconds #ใช้list
print('time of list ',t)

#ใช้ numpy
data2 = np.array(data1)

start_time1 = datetime.datetime.now()
count=np.sum(data2>50)#ไม่ใช่หาผลรวมตัวไม่เกิน แต่ที่อยุ่ในsum(เป็นเงื่อนไขT/F) sumนี้เลยนับจำนวนครั้งที่ true
end_time1 = datetime.datetime.now()
print(count)
t1 = (end_time1 - start_time1).microseconds #ใช้numpy
print('time of numpy ',t1)
````

- .microseconds จาก 0:00:00.000140 -> 140

````python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

f=open('ChulaEngineering_all_messages.txt')
tokens=f.read().split()
f.close()

print(tokens)
Counter(tokens).most_common()
````

````python
#เอาความถี่คำ ไป plot graph
Counter(tokens).values() #หาความถี่ ได้เป็น dict_values
list(Counter(tokens).values()) #ทำเป็นlist
frequencies=np.array(list(Counter(tokens).values())) #เก็บในตัวแปร frequencies

#plot
plt.subplot(211)
plt.grid(True)
plt.hist(frequencies,bins=np.arange(0,100,1)) #hist=histogram,bins=เหมือนถังใส่ข้อมูล

#การแจกแจงการกระจายด้วย subplot
x=np.arange(0.1,5,0.1)
y=x**(-2.3)
plt.subplot(212)
plt.plot(x,y)
plt.grid(True)
plt.show()

plt.hist(frequencies) #เอาbins()ออก จะเป็นแค่แท่งปกติ (ไม่โชว์ถี่ๆ)
````
